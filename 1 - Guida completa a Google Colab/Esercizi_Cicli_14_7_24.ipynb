ESERCIZIO N.4: PALINDROMI

Vedi video per spiegazione codice

```python
word = input()
while word != "stop":
  is_palindrome = True
  word_len = len(word)

  for i in range(word_len):
    j = word_len -1 -i

    if word[i] != word[j]:
      is_palindrome = False
      break

  if is_palindrome:
        print(word, "is a palindrome")
  else:
        print(word, "is not a palindrome")
  word = input()
```
